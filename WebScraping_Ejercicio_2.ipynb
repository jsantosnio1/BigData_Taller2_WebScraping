```python
import nltk
import csv
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
from nltk.probability import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('news_4.csv', newline='') as File:  
    reader = csv.reader(File)
    text = ""
    for row in reader:
        text += str(row)
        
       
```

```python
text
```

```python
text = text.lower()
```

```python
sentences = sent_tokenize(text)

print(sentences)
```

```python
tokenizer = RegexpTokenizer(r'\w+')
lisat = []
lisat = tokenizer.tokenize(str(sentences))
print(lisat)
```

```python
for i in lisat:
  if i == 'de' or i == 'la' or i == 'que' or i == 'en' or i == 'el' or i == 'y' or i == 'los' or i == 'a' or i == 'las' or i == 'se' or i == 'por' or i == 'del' or i == 'un' or i == 'para' or i == 'con' or i == 'una' or i == 'es' or i == 'no' or i == 'más' or i == 'su' or i == 'al' or i == 'lo' or i == 'como' or i == 'este' or i == 'también' or i == 'han' or i == 'entre' or i == 'le' or i == 'sus' or i == 'según' or i == 'o' or i == 'hacer' or i == 'sobre' or i == 'dos' or i == '1' or i == 'hay' or i == 'xa0m' or i == 'oct' or i == 'son' or i == 'está' or i == 'puede' or i == 'fue' or i == 'esta' or i == 'lea' or i == 'p' or i == 'estos' or i == 'cómo':
    lisat.remove(i)
```

```python
print(lisat)
```

```python
frec_dist = FreqDist(lisat)

print(frec_dist)
```

```python
frec_dist.most_common(20)
```

```python
plt.figure(figsize=(10,5))     
n = 10
plt.bar([ w[0] for w in frec_dist.most_common(n) ], [ w[1] for w in frec_dist.most_common(n) ], align='center')
```

```python
wordcloud = WordCloud(max_font_size = 50, max_words = 100, background_color = "white").generate(str(lisat))

plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
